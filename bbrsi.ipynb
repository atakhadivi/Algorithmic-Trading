```
# telegram bot to send signals
def telegram_bot_sendtext(bot_message):

    bot_token = ''
    bot_chatID = ''
    send_text = 'https://api.telegram.org/bot' + bot_token + '/sendMessage?chat_id=' + bot_chatID + '&parse_mode=Markdown&text=' + bot_message

    response = requests.get(send_text)

    return response.json()
```

```
# create data and bot
def get_data(symbol):
  global sell_price,assest_price,stop_loss
  # create pandas dataframe from binance
  client = Client(api_key, api_secret)
  df = client.get_historical_klines(symbol, Client.KLINE_INTERVAL_1HOUR, "1 day ago UTC")
  df = pd.DataFrame(df,columns= ['datetime', 'open', 'high', 'low', 'close', 'volume', 'Close_time','qoute assest volume','number of trades','taker buy base assest volume','taker buy qoute assest volume','ignore'])
  df['datetime'] = pd.to_datetime(df['datetime'],unit = 'ms')
  df.set_index('datetime',inplace=True)
  df.drop(['Close_time','qoute assest volume','number of trades','taker buy base assest volume','taker buy qoute assest volume','ignore'],axis=1,inplace=True)
  # candles.drop(candles.tail(1).index,inplace=True)
  # candles.drop(candles.head(1).index,inplace=True)
  # return df
  # candles.to_csv('out.csv')
  df = pd.DataFrame(data=df['close'], dtype=np.float64)
  # float_data = [float(x) for x in df['close']]
  close = df['close'].values
  # calculate rsi and bolinger band
  up, mid, low = BBANDS(close, timeperiod=21, nbdevup=2, nbdevdn=2, matype=0)
  rsi = RSI(close, timeperiod=14)
  rsi_length = len(rsi)
  bb_length = len(low)
  assest_price = df['close'].tail(1).values
  sell_price = df['close'].tail(1).values + ((df['close'].tail(1).values / 100.00)*1.20)
  stop_loss = df['close'].tail(1).values - ((df['close'].tail(1).values / 100.00)* 2.40)
  # if rsi[rsi_length - 1] < 70.00 and :
  #   return (rsi[rsi_length - 1])
  # calculate and condition for sell and buy
  if df['close'].tail(1).values > low[bb_length - 1]:
    if rsi[rsi_length - 1] > 33.00:
      if db.get(user.symbol== symbol)['rsi'] < 33.00:
        if db.get(user.symbol == symbol)['close'] < db.get(user.symbol == symbol)['bb']:
          return True
          text = "{} \n RSI: {} \n BB Low: {} \n price: {} \n Sell Price: {}".format(symbol,rsi[rsi_length - 1],low[bb_length - 1],df['close'].tail(1).values,sell_price)
          telegram_bot_sendtext(text)
  # db.insert({'symbol':symbol,'rsi':rsi[rsi_length - 1],'bb':low[bb_length - 1]})
  # update database for new data
  close_update = df['close'].tail(1).to_list()
  db.update({'rsi':rsi[rsi_length - 1],'bb':low[bb_length - 1],'close':close_update[0]}, user.symbol == symbol)
  # if df['close'].tail(1) < low[bb_length - 1] and rsi[rsi_length - 1] < 30:
  #   return(rsi.tail(1))
```

```
def cci30():
  session = Session()

  # HEAD requests ask for *just* the headers, which is all you need to grab the
  # session cookie
  session.head('https://cci30.com/')

  response = session.post(
      url='https://cci30.com/ajax/getConstituentsDivProd.php',
      headers={
          'Accept':
          '*/*',
          'Accept-Encoding':
          'gzip, deflate, br',
          'Accept-Language':
          'en-US,en;q=0.9,fa;q=0.8',
          'Connection':
          'keep-alive',
          'Cookie':
          '_ga=GA1.2.1811726293.1572596970; _gid=GA1.2.1592041821.1572596970; SSID_cci30=1362904212672; SV_cci30=1572596966000; SN_cci30=0%G%G1%G1572596966000%G1572596966000%G',
          'Host':
          'cci30.com',
          'Referer':
          'https://cci30.com/',
          'Sec-Fetch-Mode':
          'cors',
          'Sec-Fetch-Site':
          'same-origin',
          'User-Agent':
          'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.70 Safari/537.36',
          'X-Requested-With':
          'XMLHttpRequest'
      })

  cci30 = float(response.json()['cci30DayDiffDiv'][0:5])

  return(cci30)
```
