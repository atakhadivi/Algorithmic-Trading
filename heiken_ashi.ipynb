```python
def ichimoku(symbol,hist_time,past):
  client = Client(api_key, api_secret)
  d = client.get_historical_klines(symbol,hist_time,past)
  d = pd.DataFrame(d,columns= ['date', 'open', 'high', 'low', 'close', 'volume', 'Close_time','qoute assest volume','number of trades','taker buy base assest volume','taker buy qoute assest volume','ignore'])
  d['open'] = d['open'].astype('float')
  d['close'] = d['close'].astype('float')
  d['high'] = d['high'].astype('float')
  d['low'] = d['low'].astype('float')
  ha_close = (d['open'] + d['close'] + d['high'] + d['low']) / 4
  ha_open = [(d['open'].iloc[0] + d['close'].iloc[0]) / 2]
  for close in ha_close[:-1]:
      ha_open.append((ha_open[-1] + close) / 2)
  ha_open = np.array(ha_open)
  elements = d['high'], d['low'], ha_open, ha_close
  ha_high, ha_low = np.vstack(elements).max(axis=0), np.vstack(elements).min(axis=0)
  nine_period_high = d['high'].rolling(window= 9).max()
  nine_period_low = d['low'].rolling(window= 9).min()
  d['tenkan_sen'] = (nine_period_high + nine_period_low) /2
  # Kijun-sen (Base Line): (26-period high + 26-period low)/2))
  period26_high = d['high'].rolling(window=24).max()
  period26_low = d['low'].rolling(window=24).min()
  d['kijun_sen'] = (period26_high + period26_low) / 2
  # Senkou Span A (Leading Span A): (Conversion Line + Base Line)/2))
  d['senkou_span_a'] = ((d['tenkan_sen'] + d['kijun_sen']) / 2).shift(24)
  d['senkou_span_a_future'] = ((d['tenkan_sen'] + d['kijun_sen']) / 2)

  # Senkou Span B (Leading Span B): (52-period high + 52-period low)/2))
  period52_high = d['high'].rolling(window=51).max()
  period52_low = d['low'].rolling(window=51).min()
  d['senkou_span_b'] = ((period52_high + period52_low) / 2).shift(24)
  d['senkou_span_b_future'] = ((period52_high + period52_low) / 2)

  # The most current closing price plotted 26 time periods behind (optional)
  d['chikou_span'] = d['close'].shift(-24)
  d = pd.DataFrame(data=d, dtype=np.float64)
  open = d['open'].values
  close = d['close'].values
  high = d['high'].values
  low = d['low'].values
  tenkan_sen = d['tenkan_sen'].values
  kijun_sen = d['kijun_sen'].values
  senkou_span_a = d['senkou_span_a'].values
  senkou_span_a_future = d['senkou_span_a_future'].values
  senkou_span_b = d['senkou_span_b'].values
  senkou_span_b_future = d['senkou_span_b_future'].values
  chikou_span = d['chikou_span'].values
  pric = d['close'].values
  # print(chikou_span[-27])
  # haopen = d['haopen'].values
  # haclose = d['haclose'].values
  # hahigh = d['hahigh'].values
  # halow = d['halow'].values
  return (tenkan_sen,kijun_sen,senkou_span_a,senkou_span_a_future,senkou_span_b,senkou_span_b_future,chikou_span,pric,open,high,low,ha_open,ha_close,ha_high,ha_low)
```
