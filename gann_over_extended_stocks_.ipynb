```python
import yfinance as yf
from datetime import datetime, timedelta
import requests
import pandas as pd
import numpy as np
import time
from csv import reader
import math as Math
from pandas_datareader import data
import talib as ta
```

```python
with open('data.csv', 'r') as read_obj:
    csv_reader = reader(read_obj)
    for i in csv_reader:
      try:
        data = yf.download(i,start = "2021-08-01", end= "2021-09-01")
        close = data['Close'].values
        A = ta.MA(close, timeperiod=9, matype=0)
        cal_mas = np.sqrt(A)
        cal_over = ((np.sqrt(close) - cal_mas) * 360) / 2
        cal_over_fin = ta.MA(cal_over, timeperiod=2, matype=0)
        if cal_over_fin[-2] > 180 and cal_over_fin[-1] < 180:
          print("sell", i)
        elif cal_over_fin[-2] < -180 and cal_over_fin[-1] > -180:
          print("buy", i)
      except:
        pass
```
